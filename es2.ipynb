### DATASET PREPARATION


```python
import pandas as pd
import numpy  as np
from random import choice
from random import choices
from random import randint
import math

aq = pd.read_csv('datasets/air+quality/AirQualityUCI.csv', sep=';')
aq = aq[aq.columns[2:-2]].dropna()
classes = list(filter(lambda x: x[-4:] == '(GT)', aq.columns))
features = list(set(aq.columns) - set(classes))
aq = aq[features]
aq["T"] = aq["T"].apply(lambda x: float(x.replace(',', '.')))
aq["RH"] = aq["RH"].apply(lambda x: float(x.replace(',', '.')))
aq["AH"] = aq["AH"].apply(lambda x: float(x.replace(',', '.')))

aq = aq.rename(columns={ 'PT08.S1(CO)': 'CO', 
                         'PT08.S5(O3)':'OOO', 
                         'PT08.S3(NOx)': 'NOX',
                         'PT08.S2(NMHC)': 'NMHC',  
                         'PT08.S4(NO2)': 'NOO',
                         'T': 'Temperature'})
# Replace negative values (error/missing) with -1
for c in aq.columns:
    aq[c] = aq[c].apply(lambda x: x if x >= 0 else -1)

enc = {} 
for c in aq.columns:
    enc[c] = list(set(aq[c].dropna()))
    enc[c].sort()

print(aq)

# Define A (all features except CO) and the target interval Y of interest
A = set(aq.columns) - {"CO"}
Y = {("CO", 1000.0, 1600.0)}
```

### TEST UNIVERSE GENERATION


```python
# Build the universe of tests for each variable in A.
# Each test is a condition of the form: "feature <= value" or "feature >= value"
# Each triple represents a boolean test used as a pattern condition
T = []
for a in A:
   T += [ (a, s, v ) for v in set(aq[a]) for s in ["<=",  ">="] ]
```

### CPO AND SHAPLEY


```python
from functools import lru_cache

# Compute support of Ai (the fraction of dataset rows satisfying all conditions in Ai)
# Optimized using boolean masks to avoid nested loops
def support(Ai):
    if not Ai:
        return 0
    mask = pd.Series(True, index=aq.index)
    for col, low, high in Ai:
        mask &= (aq[col] >= low) & (aq[col] <= high)  # Vectorized operation much faster than row-wise loops
    return mask.mean()  # Return the fraction of rows satisfying Ai

# Cache support results to avoid recomputing the same sets
@lru_cache(maxsize=50000)
def cached_support(frozen_Ai):
    return support(set(frozen_Ai))

# Define support for interval Y (CO between 1000.0 and 1600.0)
support_single = support(Y)  # Compute support of Y only once

# Compute J-measure to evaluate how well Ai predicts Y
# Use support with and without Y to estimate the information provided by adding Y
def cpo(Ai):
    Ai_frozen = frozenset(Ai)
    Ai_union_frozen = frozenset(Ai | Y)

    s_ai = cached_support(Ai_frozen)
    s_ai_union = cached_support(Ai_union_frozen)

    if s_ai == 0 or s_ai_union == 0:
        return 0

    conf = s_ai_union / s_ai  # rule confidence: how often Y is true given Ai

    # The J-measure formula changes depending on whether confidence is maximal
    if conf == 1:
        return s_ai * conf * math.log2(s_ai_union / (s_ai * support_single))
    else:
        return s_ai * (
            conf * math.log2(s_ai_union / (s_ai * support_single)) +
            (1 - conf) * math.log2((1 - conf) / (1 - support_single))
        )

# Shapley value computation: evaluates Y marginal contribution to prediction,
# averaging over all possible condition combinations generated from T
def shapley(T, Y, n_sample):
    F = set(T) - {Y}  # Remove Y from the predictor universe
    F_list = list(F)
    seen = set()
    subsets = []

    # Generate n_sample random subsets of F without duplicates
    while len(subsets) < n_sample:
        Xi = frozenset(choices(F_list, k=randint(0, len(F_list))))
        if Xi not in seen:
            seen.add(Xi)
            subsets.append(Xi)

    N = 0.0 # Weighted sum of marginal contributions
    D = 0.0 # Sum of weights

    F_set = set(F) # Used to compute complements

    for Xi in subsets:
        len_x = len(Xi)
        Di = math.factorial(len_x) * math.factorial(len(F_set) - len_x)  # Combinatorial weight for Shapley

        Xi_set = set(Xi)
        complement_set = F_set - Xi_set

        # Compute Y marginal contribution on Xi and on its complement
        N += Di * (cpo(Xi_set | {Y}) - cpo(Xi_set))
        N += Di * (cpo(complement_set | {Y}) - cpo(complement_set))
        D += 2 * Di

    return N / D if D != 0 else 0
```

### AI CREATION AND NORMALIZATION


```python
# Iterative search for a valid AI set (i.e., cpo != 0)
# Normalize 10 random tests into valid intervals before computing cpo

while True:
    randomico = choices(T, k=10)  # Randomly sample 10 tests (threshold conditions on attributes)
    ts = randomico
    AI = []

    # Normalization phase: for each extracted test, build a consistent [min, max] interval
    for test in ts:
        if test[1] == '<=':
            # Search for the opposite test ('>=') sullo stesso attributo con valore coerente
            wit = [t for t in ts if t[1] == '>=' and t[0] == test[0] and t[2] <= test[2]]
            if len(wit) == 0:
                # If no compatible test exists, use the attribute minimum as interval start
                AI.append((test[0], min(set(aq[test[0]]) - {-1, 0}), test[2]))
            else:
                # Otherwise, use the maximum compatible value to build the interval
                AI.append((test[0], max(wit, key=lambda x: x[2])[2], test[2]))
        else:
            wit = [t for t in ts if t[1] == '<=' and t[0] == test[0] and t[2] >= test[2]]
            if len(wit) == 0:
                AI.append((test[0], test[2], max(set(aq[test[0]]))))
            else:
                AI.append((test[0], test[2], max(wit, key=lambda x: x[2])[2]))

    AI = set(AI)

    # Keep searching until an AI set with cpo > 0 is found
    if cpo(AI) > 0:
        break
```

```python
# Compute Shapley contribution for each interval in AI
shapley_scores = {}
total_shapley = 0.0

for rule in AI:
    shap_value = shapley(AI, rule, 100)
    shapley_scores[rule] = shap_value
    total_shapley += shap_value

# Print results
for rule, value in shapley_scores.items():
    print(f"{rule} : {value:.4f}")

print(f"CPO dell'insieme AI: {cpo(AI):.4f}")
print(f"Sommatoria dei singoli valori di Shapley: {total_shapley:.4f}")
```
