This notebook implements and analyzes a stochastic scheduling system as described in Exercise 3. The project is organized into the following steps:

b. Implementation of the system and its evolution logic (step function).
c. Learning of a Markov Decision Process (MDP) from system behavior using AALpy.

### a) Design of a Stochastic Schedule Example

This section defines the components of the scheduling problem Sch = (T, C, Δ, π), as requested by the assignment. A non-trivial example is used, with stochastic durations and flexible constraints, to make the following analysis meaningful. The existence of at least one winning plan guarantees that the problem is solvable.


```python
# === a) Scheduling Problem Definition ===

# 1. Set of tasks T to schedule.
tasks = {'T1', 'T2', 'T3'}

# 2. Definition of duration intervals (Î”) per ogni task.
# Stochastic durations for T1 and T2 introduce uncertainty into the system.
duration_intervals = {
    'T1': (2, 3),  # Stochastic duration
    'T2': (1, 3),  # Wider stochastic range
    'T3': (2, 2)   # Fixed duration to show a mixed case
}

# 3. Probability distribution (Ï€) per ogni possibile durata.
# This defines the stochastic nature of task behavior.
probabilities = {
    ('T1', 2): 0.6,
    ('T1', 3): 0.4,
    ('T2', 1): 0.2,
    ('T2', 2): 0.5,
    ('T2', 3): 0.3,
    ('T3', 2): 1.0
}

# 4. Set of temporal constraints (C) che legano i task tra loro.
# Constraints are defined as (Task1, accessor1, min_diff, max_diff, Task2, accessor2).
constraints = {
    # T2 must start between 1 e 3 time units after the end of T1.
    ('T1', 'e', 1, 3, 'T2', 'b'),

    # T3 must start between 2 e 4 time units after the end of T2.
    ('T2', 'e', 2, 4, 'T3', 'b')
}

# A winning plan is an example of a valid schedule satisfying all constraints.
# Its existence is a key sanity check confirming that the problem is solvable.
print("--- Stochastic Schedule Example --- ")
print(f"Tasks: {tasks}")
print(f"Constraints: {constraints}")
print("Esempio di winning plan: {'T1': [0, 2], 'T2': [3, 5], 'T3': [7, 9]}")
```

### b) Stochastic System Implementation

This section implements the SchedulingSystem class, which models the stochastic system described in Section 3 of the assignment. The class manages the state (partial plan, current time) and the transition function (execute_step) that drives system evolution.


```python
# === b) Stochastic System Implementation ===

import random

# This class encapsulates state and logic of the stochastic scheduling system.
class SchedulingSystem:
    def __init__(self, tasks, constraints, duration_intervals, probabilities):
        self.tasks = tasks
        self.constraints = constraints
        self.duration_intervals = duration_intervals
        self.probabilities = probabilities
        self.plan = {}
        self.current_time = 0

    # Reset the system to its initial state (empty plan, time 0).
    # This is crucial for the learning algorithm, which executes multiple independent queries.
    def reset(self):
        self.plan = {}
        self.current_time = 0

    # Utility functions to access interval start (b) and end (e).
    def get_start(self, interval): return interval[0]
    def get_end(self, interval): return interval[1]
    
    # Check whether the current plan violates any temporal constraint.
    # Return True if at least one constraint is violated, otherwise False.
    def check_constraints(self):
        for c in self.constraints:
            task1, acc1, min_diff, max_diff, task2, acc2 = c
            if task1 in self.plan and task2 in self.plan:
                val1 = self.get_start(self.plan[task1]) if acc1 == 'b' else self.get_end(self.plan[task1])
                val2 = self.get_start(self.plan[task2]) if acc2 == 'b' else self.get_end(self.plan[task2])
                diff = val2 - val1
                if not (min_diff <= diff <= max_diff):
                    return True
        return False

    # Simulate system stochasticity by sampling task duration
    # according to its probability distribution (Ï€).
    def choose_duration(self, task):
        durations = list(range(self.duration_intervals[task][0], self.duration_intervals[task][1] + 1))
        probs = [self.probabilities.get((task, d), 0) for d in durations]
        return random.choices(durations, weights=probs, k=1)[0]

    # Implement the logic of the `step` function (Algorithm 1).
    # Take a task as input, update the plan, and check consistency
    # and advance time. The output ('fail', 'success', 0, 1) represents 
    # the observation used by the learning algorithm.
    def execute_step(self, task_to_schedule):
        if self.check_constraints():
            return "fail"
        if len(self.plan) == len(self.tasks):
            return "success"
            
        if task_to_schedule not in self.plan:
            duration = self.choose_duration(task_to_schedule)
            self.plan[task_to_schedule] = (self.current_time, self.current_time + duration)
            if self.check_constraints():
                return "fail"
            if len(self.plan) == len(self.tasks):
                return "success"
        
        self.current_time += 1
        
        if task_to_schedule in self.plan:
            if self.current_time > self.get_end(self.plan[task_to_schedule]):
                return 1 # Output for completed task
            else:
                return 0 # Output for running task
        return 0
```

### c) MDP Extraction via Automata Learning

In this phase, SchedulingSystem is treated as a black box and its behavior is learned as a Markov Decision Process (MDP). The AALpy library is used, which provides active automata learning algorithms for stochastic models. The result is a formal and analyzable model of the system.


```python
# === c) MDP Extraction ===
from aalpy.base import SUL 
from aalpy.learning_algs import run_stochastic_Lstar 
from aalpy.oracles import RandomWalkEqOracle
from aalpy.utils import visualize_automaton

# Wrapper class making the system compatible with the AALpy interface.
# This adapter follows the SUL (System Under Learning) pattern.
class SchedulingSUL(SUL):
    def __init__(self, tasks, constraints, duration_intervals, probabilities):
        super().__init__()
        self.scheduling_system = SchedulingSystem(tasks, constraints, duration_intervals, probabilities)
        self.alphabet = list(tasks) # Input alphabet is the set of tasks

    def pre(self):
        self.scheduling_system.reset()

    def post(self):
        pass

    def step(self, letter):
        if letter is None:
            return None
        return self.scheduling_system.execute_step(letter)

# This function orchestrates the learning process.
# It uses `run_stochastic_Lstar`, an implementation of the L* algorithm for stochastic systems.
# The oracle (`RandomWalkEqOracle`) tests the current hypothesis to find counterexamples.
def learn_mdp_from_system(scheduling_sul):
    alphabet = scheduling_sul.alphabet
    
    eq_oracle = RandomWalkEqOracle(alphabet, sul=scheduling_sul, num_steps=15000, reset_prob=0.1)

    learned_mdp = run_stochastic_Lstar(alphabet, scheduling_sul, eq_oracle, n_c=3, n_resample=30, print_level=1)
    
    return learned_mdp

# Run the learning process
print("Avvio dell'apprendimento dell'MDP...")
sul_instance = SchedulingSUL(tasks, constraints, duration_intervals, probabilities)
learned_mdp = learn_mdp_from_system(sul_instance)

# The final result is an MDP that is visualized and saved.
# This model is a formal representation of system behavior.
print("\nApprendimento completato. Visualizzazione dell'MDP...")
visualize_automaton(learned_mdp, path='learned_mdp.pdf')
print("MDP salvato in 'learned_mdp.pdf'")
```

### d) Fair Strategy Implementation

With the MDP available, a control strategy can be derived. The assignment requires a fair strategy. Here, an **optimal policy** is computed using the **Value Iteration** algorithm. This policy maximizes the probability of reaching a success state from each possible state, satisfying and strengthening the fairness requirement.


```python
# === d) Fair Strategy Implementation ===
import networkx as nx
from networkx.drawing.nx_pydot import to_pydot

# Compute an optimal policy for the given MDP using Value Iteration.
# The goal is to maximize the probability of reaching a state labeled 'success'.
def find_optimal_policy(mdp):
    # Initialization: success probability is 1 for state 'success', 0 for others.
    prob_success = {s: 1.0 if s.output == 'success' else 0.0 for s in mdp.states}
    
    # The algorithm iteratively propagates probability values backward through the MDP
    # until convergence (when values no longer change).
    for _ in range(len(mdp.states) * 2): # Number of iterations sufficient for convergence
        updated = False
        for state in mdp.states:
            if state.output in {'success', 'fail'}: continue

            # For the current state, compute the maximum value by choosing the best action.
            max_prob = 0.0
            for action in mdp.get_input_alphabet():
                # The value of an action is the weighted sum of successor success probabilities.
                current_action_prob = sum(prob * prob_success[next_state] for next_state, prob in state.transitions.get(action, []))
                max_prob = max(max_prob, current_action_prob)
            
            if prob_success[state] != max_prob:
                prob_success[state] = max_prob
                updated = True
        if not updated: break # Early exit when convergence is reached

    # Once the value function (prob_success) is stable, extract the optimal policy.
    # For each state, choose the action maximizing expected value.
    optimal_policy = {}
    for state in mdp.states:
        if state.output in {'success', 'fail'}: continue
            
        best_action, best_prob = None, -1.0
        for action in mdp.get_input_alphabet():
            current_action_prob = sum(prob * prob_success[next_state] for next_state, prob in state.transitions.get(action, []))
            if current_action_prob > best_prob:
                best_prob, best_action = current_action_prob, action
        optimal_policy[state] = best_action
        
    return optimal_policy

# Visualize the computed policy as a subgraph of the original MDP.
# Optimal actions are highlighted for easier interpretation.
def visualize_policy(mdp, policy, file_name="fair_strategy.pdf"):
    policy_graph = nx.MultiDiGraph()
    for state in mdp.states:
        label = f"{state.state_id}\n({state.output})"
        color = 'lightgreen' if state.output == 'success' else 'lightcoral' if state.output == 'fail' else 'lightblue'
        policy_graph.add_node(state.state_id, label=label, style='filled', fillcolor=color)

    for state, action in policy.items():
        if action is not None and action in state.transitions:
            for next_state, prob in state.transitions[action]:
                label = f"{action} ({prob:.2f})"
                policy_graph.add_edge(state.state_id, next_state.state_id, label=label, color='red', penwidth=2.0)

    pydot_graph = to_pydot(policy_graph)
    pydot_graph.write_pdf(file_name)
    print(f"Policy visualizzata e salvata in {file_name}")

# Compute and visualize policy
print("\nComputing fair strategy (optimal policy) via Value Iteration...")
fair_policy = find_optimal_policy(learned_mdp)
visualize_policy(learned_mdp, fair_policy)
```

### e) Trace Generation

Using the optimal policy, system behavior is simulated to generate an event log. Each trace in the log represents one execution of the scheduling process guided by policy decisions. This log is the input for the next process mining phase.


```python
# === e) Trace Generation ===
from pm4py.objects.log.obj import EventLog, Trace, Event
from pm4py.objects.log.exporter.xes import exporter as xes_exporter
from datetime import datetime

# Generate a single trace (action sequence) by simulating the MDP with the given policy.
# Simulation starts from the initial state and continues by selecting policy actions
# and stochastic MDP transitions.
def generate_trace(mdp, policy, max_steps=100):
    trace_log = []
    current_state = mdp.initial_state
    
    steps = 0
    while current_state.output not in {'success', 'fail'} and steps < max_steps:
        action = policy.get(current_state)
        if action is None: break
        
        trace_log.append(action)
        
        transitions = current_state.transitions.get(action, [])
        if not transitions: break
        
        # The next state is sampled stochastically according to MDP probabilities.
        next_states = [t[0] for t in transitions]
        probabilities = [t[1] for t in transitions]
        current_state = random.choices(next_states, weights=probabilities, k=1)[0]
        steps += 1
    
    return trace_log

# Generate a full log by running simulation for the specified number of traces.
def generate_log_of_traces(mdp, policy, num_traces=1000):
    print(f"\nGenerating {num_traces} traces...")
    log = [generate_trace(mdp, policy) for _ in range(num_traces)]
    print("Generation completed.")
    return log

# Convert the trace list to standard XES format (eXtensible Event Stream)
# and save it to file. XES is the input format for most process mining tools.
def export_log_to_xes(log_data, filename="generated_log.xes"):
    log = EventLog()
    for i, trace_list in enumerate(log_data):
        trace = Trace()
        trace.attributes["concept:name"] = f"Case_{i+1}"
        timestamp = datetime.now()
        
        for event_name in trace_list:
            event = Event()
            event["concept:name"] = event_name
            event["time:timestamp"] = timestamp
            trace.append(event)
            
        log.append(trace)
        
    xes_exporter.apply(log, filename)
    print(f"Log exported in XES format to file: {filename}")

# Generate and save log
generated_traces = generate_log_of_traces(learned_mdp, fair_policy, num_traces=1000)
export_log_to_xes(generated_traces, "generated_log.xes")
```

### f) BPMN/Petri Net Mining

The final phase applies process mining techniques to the generated event log. Using the pm4py library, a process model (a Petri net) is discovered to describe the recorded behavior. Finally, model quality is evaluated through a conformance metric (fitness) to measure how faithfully it reproduces the log.


```python
# === f) BPMN/Petri Net Mining ===
import pm4py
from pm4py.objects.log.importer.xes import importer as xes_importer
from IPython.display import Image, display

# This function executes the full process mining cycle: log loading,
# model discovery, and conformance evaluation.
def mine_and_visualize_process(log_path, petri_net_path="discovered_petri_net.png"):
    # 1. Load the event log from the XES file.
    log = xes_importer.apply(log_path)
    
    # 2. Discover the process model using the Inductive Miner.
    # This algorithm is effective in generating well-structured models (e.g., sound Petri nets).
    net, initial_marking, final_marking = pm4py.discover_petri_net_inductive(log)
    
    # 3. Visualize the discovered Petri net.
    # This model provides a high-level view of process control flow.
    pm4py.save_vis_petri_net(net, initial_marking, final_marking, petri_net_path)
    print(f"\n=== Rete di Petri Scoperta (salvata in {petri_net_path}) ===")
    display(Image(filename=petri_net_path))
    
    # 4. Evaluate model quality via conformance checking.
    # The 'fitness' metric measures model ability to reproduce log traces.
    # A value close to 1.0 indicates an excellent match.
    fitness = pm4py.fitness_alignments(log, net, initial_marking, final_marking)
    print("\n--- Conformance Analysis ---")
    print(f"  - Fitness del modello (alignments): {fitness['average_trace_fitness']:.4f}")
    
    return net, initial_marking, final_marking

# Run process mining analysis
print("\nAvvio del Process Mining sul log generato...")
mined_net, im, fm = mine_and_visualize_process("generated_log.xes")
```
