```python
def load_arff_data(file_path: str, label_map: Dict[int, int] = None) -> Tuple[pd.DataFrame, pd.Series]:
    """Load an ARFF dataset, separating features from labels."""
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()
    
    # Find and parse the start of the data section
    data_start_line = next(i for i, line in enumerate(lines) if "@data" in line.lower()) + 1
    data = [list(map(float, line.strip().split(","))) for line in lines[data_start_line:]]

    # Separate features (all columns except the last) from labels (last column)
    df = pd.DataFrame([row[:-1] for row in data])
    labels = pd.Series([int(row[-1]) for row in data], dtype=int)
    
    # Apply label mapping if provided (e.g., to standardize -1 to 0)
    if label_map:
        labels.replace(label_map, inplace=True)
        
    return df, labels
```

## 2. Model Implementation


### Sequential Boost Classifier (SBC)


```python
@numba.jit(nopython=True, fastmath=True)
def _find_best_learner_numba(x_col: np.ndarray, y: np.ndarray, w: np.ndarray) -> Tuple[Tuple[int, float], float]:
    """Trova lo stump (soglia e direzione) ottimale per una singola feature cercando la soglia che meglio separa i dati minimizzando l'errore pesato."""
    best_err = 1.0
    best_d = 0
    best_th = 0.0
    
    # Iterate over all possible thresholds (unique feature values)
    thresholds = np.unique(x_col)
    for th in thresholds:
        # Compute error for direction d=1
        err_d1 = 0.0
        for i in range(len(x_col)):
            pred = 1 if (x_col[i] - th) >= 0 else -1
            if pred != y[i]:
                err_d1 += w[i]
        
        # Update best stump if current error is lower
        if err_d1 < best_err:
            best_err = err_d1
            best_d = 1
            best_th = th
        
        # Evaluate opposite direction (d=-1) without recomputing everything
        err_d_neg1 = 1.0 - err_d1
        if err_d_neg1 < best_err:
            best_err = err_d_neg1
            best_d = -1
            best_th = th
            
    return (best_d, best_th), best_err

class SBC:
    """Implements the base Sequential Boost Classifier (SBC)."""
    def __init__(self):
        """Initializes the classifier, storing a list of weak learners (wl)."""
        self.wl: List[Tuple[float, Tuple[int, float], int]] = []
        self.tot_weight: float = 0.0

    def fit(self, x: pd.DataFrame, y: pd.Series, n_weak_learners: int = -1):
        """Trains the sequential classifier using an AdaBoost-like approach."""
        if n_weak_learners == -1:
            n_weak_learners = x.shape[1] # One weak learner per feature by default

        # Initialize sample weights uniformly
        n_samples = len(x)
        weights = np.full(n_samples, 1 / n_samples)
        x_np = x.to_numpy()
        y_np = y.to_numpy()

        # Train a weak learner (stump) for each feature
        for i in range(n_weak_learners):
            # Find best stump for current feature
            h, err = _find_best_learner_numba(x_np[:, i], y_np, weights)
            if err >= 0.5 - 1e-10: continue # Ignore classifiers with random-level or worse performance
            
            # Compute classifier weight (alpha) from its error (accuracy)
            alpha = 0.5 * np.log((1 - err) / (err + 1e-10))
            self.tot_weight += alpha

            # Update sample weights
            predictions = h[0] * self.sign(x_np[:, i] - h[1]) # predictions are -1 or 1
            weights *= np.exp(-alpha * y_np * predictions)
            weights /= np.sum(weights)

            # Store trained weak learner, its weight, and feature index
            self.wl.append((alpha, h, i))
        return self

    def predict(self, x: pd.DataFrame) -> np.ndarray:
        """Performs prediction by aggregating weighted votes from all weak learners."""
        scores = np.zeros(len(x))
        x_np = x.to_numpy()
        for alpha, h, feature_idx in self.wl:
            d, th = h
            scores += alpha * d * self.sign(x_np[:, feature_idx] - th)
        return self.sign(scores)
    
    @staticmethod
    def sign(s: np.ndarray) -> np.ndarray:
        """Utility function to compute the sign of an array, returning 1 or -1."""
        return np.where(s >= 0, 1, -1)
```

### Bagging Ensemble of Sequential Boost Classifiers


```python
def _train_single_be_sbc(args: Tuple) -> Dict:
    """Funzione helper per addestrare un singolo SBC in modalitÃ  One-vs-All.
    Returns the model and its validation performance (accuracy, FNR)."""
    x_train, y_train, x_val, y_val, target_class = args
    
    # Create binary labels (1 for target class, -1 for others)
    y_train_ovr = np.where(y_train == target_class, 1, -1)
    y_val_ovr = np.where(y_val == target_class, 1, -1)
    
    # Train SBC and compute its performance for vote weighting
    sbc = SBC().fit(x_train, pd.Series(y_train_ovr))
    predictions_val = sbc.predict(x_val)
    accuracy = np.mean(predictions_val == y_val_ovr)
    
    positives_in_val = (y_val_ovr == 1)
    fnr_denominator = np.sum(positives_in_val)
    fnr = np.sum(predictions_val[positives_in_val] == -1) / fnr_denominator if fnr_denominator > 0 else 0.0
    
    return {'model': sbc, 'class': target_class, 'accuracy': accuracy, 'fnr': fnr}

class SBC_BE:
    """Implements a Bagging Ensemble of SBC classifiers (BE-SBC) for multiclass problems."""
    def __init__(self, n_estimators: int, vote_mode: str = "trackRecord", n_jobs: int = -1):
        """Initializes the supervised ensemble by setting estimator count and voting mode."""
        self.n_estimators = n_estimators
        self.vote_mode = vote_mode
        self.n_jobs = n_jobs
        self.ensemble: List[Dict[str, Any]] = []
        self.classes_ = None

    def fit(self, x_train: pd.DataFrame, y_train: pd.Series, x_val: pd.DataFrame, y_val: pd.Series):
        """Trains a set of SBC classifiers in parallel, each specialized on one class."""
        self.classes_ = np.sort(y_train.unique())
        
        # Create parallel training tasks, assigning each a random target class
        tasks = []
        for _ in range(self.n_estimators):
            target_class = np.random.choice(self.classes_)
            tasks.append((x_train, y_train.to_numpy(), x_val, y_val.to_numpy(), target_class))
            
        # Run training tasks in parallel using Joblib
        self.ensemble = Parallel(n_jobs=self.n_jobs)(delayed(_train_single_be_sbc)(task) for task in tasks)
        return self

    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """Computes class probabilities for input samples by aggregating ensemble votes."""
        X_df = pd.DataFrame(X)
        all_probas = []

        # For each sample to predict
        for _, row in X_df.iterrows():
            class_scores = {c: 0.0 for c in self.classes_}
            total_weight = 0.0

            # collect votes from each classifier in the ensemble.
            for sbc_info in self.ensemble:
                pred = sbc_info['model'].predict(pd.DataFrame([row]))[0]
                
                # Assign vote weight based on selected mode 
                if self.vote_mode == "trackRecord":
                    weight = sbc_info['accuracy'] if pred == 1 else sbc_info['fnr']
                elif self.vote_mode == "weightVoting":
                    weight = sbc_info['accuracy'] if pred == 1 else 0.0
                else: # majorVoting
                    weight = 1.0 if pred == 1 else 0.0

                class_scores[sbc_info['class']] += weight
                total_weight += weight

            # Normalize total scores to obtain a probability distribution.
            if total_weight > 0:
                row_probas = [class_scores[c] / total_weight for c in self.classes_]
            else:
                row_probas = [1.0 / len(self.classes_)] * len(self.classes_)
                
            all_probas.append(row_probas)

        return np.array(all_probas)
```

### Isolation-based Sequential Boost Classifier


```python
def _train_single_isbc(x: pd.DataFrame) -> SBC:
    """Trains a single SBC to distinguish in-bag samples (label +1) from out-of-bag (-1)."""
    n_samples = len(x)
    
    # Create a bootstrap sample and identify out-of-bag samples
    bootstrap_indices = np.random.choice(x.index, size=n_samples, replace=True)
    oob_indices = x.index.difference(np.unique(bootstrap_indices))
    
    # Prepare artificial training set for the unsupervised approach
    x_in_bag = x.loc[bootstrap_indices]
    x_oob = x.loc[oob_indices]
    x_train = pd.concat([x_in_bag, x_oob], ignore_index=True)
    y_train = pd.Series(np.hstack([np.ones(len(x_in_bag)), np.full(len(x_oob), -1)]))

    sbc = SBC().fit(x_train, y_train)
    return sbc
    
class I_SBC:
    """Implements an unsupervised I-SBC ensemble used to compute distances between time series. Measures similarity."""
    def __init__(self, n_estimators: int = 100, n_jobs: int = -1):
        """Initializes the unsupervised ensemble."""
        self.n_estimators = n_estimators
        self.n_jobs = n_jobs
        self.ensemble: List[SBC] = []

    def fit(self, x: pd.DataFrame):
        """Trains `n_estimators` SBC models in parallel on random bootstrap samples."""
        self.ensemble = Parallel(n_jobs=self.n_jobs)(
            delayed(_train_single_isbc)(x) for _ in range(self.n_estimators)
        )
        return self

    def _get_all_predictions(self, x: pd.DataFrame) -> np.ndarray:
        """Collects predictions of each weak learner from each SBC for all input data."""
        max_wl = max(len(sbc.wl) for sbc in self.ensemble if sbc.wl) if self.ensemble else 0
        if max_wl == 0: return np.empty((len(self.ensemble), len(x), 0))
        
        all_preds = np.full((len(self.ensemble), len(x), max_wl), np.nan)
        x_np = x.to_numpy()

        for i, sbc in enumerate(self.ensemble):
            for j, wl_info in enumerate(sbc.wl):
                _, h, feature_idx = wl_info
                d, th = h
                all_preds[i, :, j] = d * SBC.sign(x_np[:, feature_idx] - th)
        return all_preds

    def breiman_distance_matrix(self, x: pd.DataFrame) -> np.ndarray:
        """Computes Breiman distance: two points are far apart if SBCs disagree on predictions."""
        preds = self._get_all_predictions(x)
        if preds.shape[2] == 0: return np.ones((len(x), len(x)))
        
        comparisons = (preds[:, :, np.newaxis, :] == preds[:, np.newaxis, :, :])
        comparisons = np.nan_to_num(comparisons, nan=True).astype(bool)
        all_wl_match = np.all(comparisons, axis=3)
        similarity_matrix = np.mean(all_wl_match, axis=0)
        return 1 - similarity_matrix

    def ratio_distance_matrix(self, x: pd.DataFrame) -> np.ndarray:
        """Computes Ratio distance: the fraction of mismatched decisions among all possible decisions."""
        preds = self._get_all_predictions(x)
        if preds.shape[2] == 0: return np.ones((len(x), len(x)))
        
        comparisons = (preds[:, :, np.newaxis, :] == preds[:, np.newaxis, :, :])
        num_matches = np.nansum(comparisons, axis=(0, 3))
        total_decisions = np.sum([len(sbc.wl) for sbc in self.ensemble])
        
        if total_decisions == 0: return np.ones((len(x), len(x)))
        similarity_matrix = num_matches / total_decisions
        return 1 - similarity_matrix

    def zhu_distance_matrix(self, x: pd.DataFrame) -> np.ndarray:
        """Computes Zhu distance, based on the length of the prefix of matching decisions."""
        preds = self._get_all_predictions(x)
        if preds.shape[2] == 0: return np.ones((len(x), len(x)))
        n_estimators, n_samples, n_wl = preds.shape

        match_matrix = (preds[:, :, np.newaxis, :] == preds[:, np.newaxis, :, :])
        cumulative_match = np.cumprod(match_matrix, axis=3, dtype=float)
        max_prefix_per_estimator = np.sum(cumulative_match, axis=3)

        total_similarity = 0
        for i, sbc in enumerate(self.ensemble):
            num_wl_sbc = len(sbc.wl)
            if num_wl_sbc > 0:
                total_similarity += max_prefix_per_estimator[i, :, :] / num_wl_sbc
        
        similarity_matrix = total_similarity / n_estimators if n_estimators > 0 else 0
        return 1 - similarity_matrix
```

### Isolation-based Sequential Boost Classifier - Nearest Neighbor


```python
class I_SBC_NN:
    """Implements a 1-Nearest-Neighbor (1-NN) classifier using distances computed by I-SBC."""
    def __init__(self, i_sbc_ensemble: I_SBC, distance_metric: str = "breiman"):
        """Initializes the classifier with the I-SBC ensemble and the selected distance metric."""
        self.isbc = i_sbc_ensemble
        self.distance_metric = distance_metric
        self.x_train = None
        self.y_train = None
        self.classes_ = None

    def fit(self, x_train: pd.DataFrame, y_train: pd.Series):
        """Stores the training set used to compute distances during prediction."""
        self.x_train = x_train
        self.y_train = y_train
        self.classes_ = np.sort(y_train.unique())
        return self

    def predict_proba(self, X: np.ndarray) -> np.ndarray:
        """Performs classification based on the nearest-neighbor distance for each class."""
        X_df = pd.DataFrame(X)
        all_probas = []

        # Select the specified distance function
        if self.distance_metric == "breiman":
            dist_func = self.isbc.breiman_distance_matrix
        elif self.distance_metric == "ratio":
            dist_func = self.isbc.ratio_distance_matrix
        else: # Zhu
            dist_func = self.isbc.zhu_distance_matrix

        # Compute distance matrix between test points and all training points
        full_data = pd.concat([self.x_train, X_df], ignore_index=True)
        dist_matrix_full = dist_func(full_data)
        n_train = len(self.x_train)
        dist_matrix_test_vs_train = dist_matrix_full[n_train:, :n_train]

        # For each test point
        for i in range(len(X_df)):
            distances_to_train = dist_matrix_test_vs_train[i, :]
            nearest_distances = []
            
            # find the minimum distance to training points of each class.
            for c in self.classes_:
                class_mask = (self.y_train == c)
                min_dist = np.min(distances_to_train[class_mask]) if np.any(class_mask) else 1.0
                nearest_distances.append(min_dist)
            
            # Convert minimum distances into probabilities (smaller distance = higher probability).
            scores = [1 / (d + 1e-10) for d in nearest_distances]
            denominator = np.sum(scores)
            probas = [s / denominator for s in scores] if denominator > 0 else [1/len(self.classes_)]*len(self.classes_)
            all_probas.append(probas)
            
        return np.array(all_probas)
```

```python
class MyClassifierAdapter(ClassifierAdapter):
    """Adapter that makes our models compatible with the `nonconformist` library.
    This class acts as a bridge between our implementation and the Conformal Prediction framework."""
    def __init__(self, model, x_val=None, y_val=None):
        """Initializes the adapter with the wrapped model."""
        super(MyClassifierAdapter, self).__init__(model)
        self.x_val = x_val
        self.y_val = y_val

    def fit(self, x, y):
        """Calls the proper `fit` method based on model type (SBC_BE requires a validation set)."""
        x_df = pd.DataFrame(x)
        y_s = pd.Series(y)
        
        if isinstance(self.model, I_SBC_NN):
            self.model.fit(x_df, y_s)
        elif isinstance(self.model, SBC_BE):
            if self.x_val is None or self.y_val is None:
                raise ValueError("Validation set (x_val, y_val) must be provided for SBC_BE")
            self.model.fit(x_df, y_s, self.x_val, self.y_val)

    def predict(self, x) -> np.ndarray:
        """Returns prediction probabilities, as required by `nonconformist`."""
        return self.model.predict_proba(x)
```

## 3. Analysis Pipeline and Evaluation Functions


```python
def calculate_cluster_distances(distance_matrix: np.ndarray, labels: np.ndarray, metric: str) -> float:
    """Computes average intra-cluster or inter-cluster distance from a distance matrix."""
    distances = []
    unique_labels = np.unique(labels[labels != -1])
    if len(unique_labels) < 2 and metric == 'inter': return 0.0
    if len(unique_labels) < 1: return 0.0

    if metric == 'intra':
        # Compute average distance among points within the same cluster
        for label in unique_labels:
            cluster_indices = np.where(labels == label)[0]
            if len(cluster_indices) > 1:
                rows, cols = np.triu_indices(len(cluster_indices), k=1)
                distances.extend(distance_matrix[cluster_indices[rows], cluster_indices[cols]])
    elif metric == 'inter':
        # Compute average distance among points in different clusters
        for i in range(len(unique_labels)):
            for j in range(i + 1, len(unique_labels)):
                indices1 = np.where(labels == unique_labels[i])[0]
                indices2 = np.where(labels == unique_labels[j])[0]
                if len(indices1) > 0 and len(indices2) > 0:
                    distances.extend(distance_matrix[indices1[:, None], indices2].flatten())

    return np.mean(distances) if distances else 0.0

def compute_total_entropy(cluster_labels: np.ndarray, true_labels: pd.Series) -> float:
    """Computes weighted cluster entropy, a purity measure with respect to true labels."""
    unique_cluster_ids = np.unique(cluster_labels[cluster_labels != -1])
    non_noise_points_mask = (cluster_labels != -1)
    total_points = np.sum(non_noise_points_mask)
    if total_points == 0: return 0.0
    
    total_entropy = 0
    # For each cluster, compute internal entropy and weight it by cluster size
    for cluster_id in unique_cluster_ids:
        cluster_mask = (cluster_labels == cluster_id)
        cluster_true_labels = true_labels[cluster_mask]
        
        def compute_single_cluster_entropy(labels_array: np.ndarray) -> float:
            n_points = len(labels_array)
            if n_points == 0: return 0.0
            _, counts = np.unique(labels_array, return_counts=True)
            props = counts / n_points
            return -np.sum(props * np.log2(props + 1e-10))

        cluster_entropy = compute_single_cluster_entropy(cluster_true_labels.to_numpy())
        weight = len(cluster_true_labels) / total_points
        total_entropy += weight * cluster_entropy
        
    return total_entropy
```

```python
def perform_full_analysis(
    df_train_1: pd.DataFrame, labels_train_1: pd.Series,
    df_train_2: pd.DataFrame, labels_train_2: pd.Series,
    df_test: pd.DataFrame, labels_test: pd.Series,
    dataset_name: str,
    n_estimators_isbc: int = 100,
    n_estimators_be: int = 100
):
    """Run the full analysis pipeline: clustering, BE-SBC conformal prediction, and I-SBC-NN."""
    print(f"=====================================================")
    print(f"  STARTING ANALYSIS FOR DATASET: {dataset_name.upper()}")
    print(f"=====================================================\n")

    # --- TASK 1: Clustering Analysis (I-SBC) ---
    print("\n--- TASK 1: I-SBC Clustering Analysis ---\n")
    # Train I-SBC and evaluate clustering metrics (Intra/Inter-Dist, Entropy).
    start_time = time.time()
    isbc_cluster_model = I_SBC(n_estimators=n_estimators_isbc).fit(df_train_1)
    print(f"Tempo di addestramento I-SBC ({n_estimators_isbc} estimatori): {time.time() - start_time:.2f}s")
    
    distance_matrices = {
        "Breiman-like": isbc_cluster_model.breiman_distance_matrix(df_train_1),
        "Zhu-like": isbc_cluster_model.zhu_distance_matrix(df_train_1),
        "Ratio-like": isbc_cluster_model.ratio_distance_matrix(df_train_1),
    }

    n_classes = len(labels_train_1.unique())
    cluster_results = []
    for name, matrix in distance_matrices.items():
        # Run KMeans clustering on the distance matrix to enforce a fixed number of clusters.
        kmeans = KMeans(n_clusters=n_classes, random_state=42, n_init='auto').fit(matrix)
        labels = kmeans.labels_
        
        # Compute clustering evaluation metrics.
        intra_dist = calculate_cluster_distances(matrix, labels, metric='intra')
        inter_dist = calculate_cluster_distances(matrix, labels, metric='inter')
        sep_ratio = intra_dist / (inter_dist + 1e-10)
        entropy = compute_total_entropy(labels, labels_train_1)
        
        cluster_results.append([name, f"{intra_dist:.4f}", f"{inter_dist:.4f}", f"{sep_ratio:.4f}", f"{entropy:.4f}"])

    print("\nClustering Metrics Results (with KMeans):")
    print(tabulate(cluster_results, headers=["Metric", "Intra-Dist", "Inter-Dist", "Sep. Ratio", "Entropy"], tablefmt="grid"))

    # --- TASK 2: BE-SBC Conformal Prediction ---
    print("\n--- TASK 2: BE-SBC Conformal Prediction ---\n")
    # Train and calibrate a BE-SBC conformal predictor, then evaluate error and efficiency.
    start_time = time.time()
    be_sbc_model = SBC_BE(n_estimators=n_estimators_be, vote_mode="trackRecord")
    model_adapter_be = MyClassifierAdapter(be_sbc_model, df_train_2, labels_train_2)
    icp_be = IcpClassifier(ClassifierNc(model_adapter_be))
    
    icp_be.fit(df_train_1.to_numpy(), labels_train_1.to_numpy())
    print(f"Tempo di addestramento BE-SBC ({n_estimators_be} estimatori): {time.time() - start_time:.2f}s")
    icp_be.calibrate(df_train_2.to_numpy(), labels_train_2.to_numpy())
    
    SIGNIFICANCE = 0.07
    regions_be = icp_be.predict(df_test.to_numpy(), significance=SIGNIFICANCE)
    
    efficiency_be = np.mean(np.sum(regions_be, axis=1))
    # Check for each test sample whether the conformal prediction set contains the true label.
    # Average error rate is used to verify whether the classifier is well calibrated.
    error_rate_be = 1 - np.mean([regions_be[i, labels_test.iloc[i]] for i in range(len(labels_test))])
    
    print(f"BE-SBC Conformal Prediction Results (Significance = {SIGNIFICANCE:.2f}):")
    print(f"  - Empirical Error Rate: {error_rate_be:.4f} (Obiettivo: ~{SIGNIFICANCE:.2f})")
    print(f"  - Efficienza (dimensione media del prediction set): {efficiency_be:.2f}")

    # --- TASK 3: I-SBC-NN Conformal Prediction ---
    print("\n--- TASK 3: I-SBC-NN Conformal Prediction ---\n")
    # Train and calibrate an I-SBC-NN conformal predictor, then evaluate error and efficiency.
    start_time = time.time()
    isbc_nn = I_SBC_NN(isbc_cluster_model, distance_metric="breiman")
    model_adapter_nn = MyClassifierAdapter(isbc_nn)
    icp_nn = IcpClassifier(ClassifierNc(model_adapter_nn))
    
    icp_nn.fit(df_train_1.to_numpy(), labels_train_1.to_numpy())
    print(f"Tempo di fit I-SBC-NN: {time.time() - start_time:.2f}s")
    #  The model uses the calibration set to compute nonconformity scores and learn the threshold required to guarantee
    #  the desired alpha significance level.
    icp_nn.calibrate(df_train_2.to_numpy(), labels_train_2.to_numpy())
    
    regions_nn = icp_nn.predict(df_test.to_numpy(), significance=SIGNIFICANCE)
    
    efficiency_nn = np.mean(np.sum(regions_nn, axis=1))
    error_rate_nn = 1 - np.mean([regions_nn[i, labels_test.iloc[i]] for i in range(len(labels_test))])
    
    print(f"I-SBC-NN Conformal Prediction Results (Breiman, Significance = {SIGNIFICANCE:.2f}):")
    print(f"  - Empirical Error Rate: {error_rate_nn:.4f} (Obiettivo: ~{SIGNIFICANCE:.2f})")
    print(f"  - Efficienza (dimensione media del prediction set): {efficiency_nn:.2f}")
    
    print(f"\n=====================================================")
    print(f"  END OF ANALYSIS FOR {dataset_name.upper()}")
    print(f"=====================================================\n\n")
```

### 4.1 Analysis on the ECG200 Dataset (Binary)


```python
# Load and split ECG200 into train, calibration, and test sets.
df_test_ecg, labels_test_ecg = load_arff_data("datasets/ECG200_TEST.arff", {-1: 0})
df_train_ecg, labels_train_ecg = load_arff_data("datasets/ECG200_TRAIN.arff", {-1: 0})

indices_ecg = np.random.permutation(len(df_train_ecg))
midpoint_ecg = len(indices_ecg) // 2
train_indices_ecg, cal_indices_ecg = indices_ecg[:midpoint_ecg], indices_ecg[midpoint_ecg:]

df_train_1_ecg = df_train_ecg.iloc[train_indices_ecg].reset_index(drop=True)
df_train_2_ecg = df_train_ecg.iloc[cal_indices_ecg].reset_index(drop=True)
labels_train_1_ecg = labels_train_ecg.iloc[train_indices_ecg].reset_index(drop=True)
labels_train_2_ecg = labels_train_ecg.iloc[cal_indices_ecg].reset_index(drop=True)

# Run the full analysis pipeline on ECG200.
perform_full_analysis(df_train_1_ecg, labels_train_1_ecg,
                      df_train_2_ecg, labels_train_2_ecg,
                      df_test_ecg, labels_test_ecg,
                      "ECG200",
                      n_estimators_isbc=100, 
                      n_estimators_be=100)
```
